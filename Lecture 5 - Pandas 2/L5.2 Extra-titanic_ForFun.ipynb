```python
import pandas as pd
titanic = pd.read_csv('data\\titanic.csv')
# Drop column sex:
titanic2 = titanic.drop(columns = 'Sex')
titanic2.head()
```

```python
def get_prefix_name(r):
    name = r['Name']
    comma_index = name.index(',')
    dot_index = name.index('.')
    prefix = name[comma_index+1: dot_index].strip()
    return prefix
titanic2['prefixName'] = titanic2.apply(get_prefix_name, axis = 1)
```

```python
titanic2.head()
```

```python
# get all unique prefix names:
titanic2.prefixName.unique()
```

```python
gender_dict = {
    'female': ['Mrs', 'Miss', 'Mme', 'Ms', 'Lady', 'Mlle', 'the Countess'],
    'male': ['Mr', 'Master', 'Don', 'Rev', 'Major', 'Sir', 'Col', 'Capt', 'Jonkheer']
}
```

```python
def add_gender(x):
    for gender, prefix_list in gender_dict.items():
        if x in prefix_list:
            return gender
    return None

titanic2['new_sex'] =  titanic2.prefixName.apply(add_gender)
```
